```python
# Replacing the values in the'income_range' column with the class marks of the intervals. 

ese_hi = df['income_range'].replace('Less than ₹12500', 12500)
ese_hi = list(ese_hi.replace('₹100000 or More', 100000))
# Spliting & converting the end points to integers.
from pandas.core.arrays.sparse import dtype
sjnsdkj = []
for i in ese_hi:
  if type(i) == str:
    sjnsdkj.append(i.split(' - '))
  else:
    sjnsdkj.append(i)
sjnsdkj[:10]
# Taking the average
income_avg = []
for i in sjnsdkj:
  avg = 0
  if type(i) == list:
    avg = (int(i[0][1:]) + int(i[1][1:]))/2
    income_avg.append(avg)
  else:
    income_avg.append(i)
income_avg[:10]
# Updating the series
df['income_range'] = income_avg
```

**Converting other string data entries into int or float values:**

1) 'less1' to 1, 'gt8' to 8, 'never' to 0, intervals to their class mark.

2) '50plus' to 50, 'below20' to 20.

3) '2days' to 2*24, '10hours' to 10.

```python
# Testing different models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import xgboost

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score
```

```python
# Encoding the categorical values.
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
```

```python
from sklearn.compose import make_column_transformer
col_trans = make_column_transformer((ohe,
                                    ['Marital Status', 'restaurant type', 'gender', 'car', 'Customer type', 'Qualification', 'Job/Job Industry', 'Climate', 'drop location']),
                                    remainder = 'passthrough')
X = col_trans.fit_transform(X)
```

**Model: RandomizedSearchCV & GridSearchCV with xgboost.XGBClassifier**

```python
# Setting parameter dictionary for hyperparameter tuning
learning_rate_list = [0.05,0.10,0.15,0.20,0.25,0.30]
max_depth_list = [ 3, 4, 5, 6, 8, 10, 12, 15]
n_estimators_list = [3000, 4000, 5000, 6000, 7000]

params_dict = {"learning_rate": learning_rate_list,
               "max_depth": max_depth_list,
               "n_estimators": n_estimators_list}
```

```python
# RandomizedSearchCV
model_xgb1 = RandomizedSearchCV(estimator = xgboost.XGBClassifier(use_label_encoder = False,
                                                             random_state = 42),
                         param_distributions = params_dict,
                         n_iter = 5,
                         scoring = 'roc_auc',
                         cv = 5,
                         n_jobs = -1,       
                         return_train_score = True)
model_xgb1.fit(X_train, y_train)
```

```python
# Getting the best parameters from RandomSearchCV
print('params:', model_xgb1.best_params_)
```

```python
# Updating parameter dictionary for hyperparameter tuning
learning_rate_list = [0.02,0.05,0.10]
max_depth_list = [3,4,5]
n_estimators_list = [3000,4000,5000]

params_dict = {"learning_rate": learning_rate_list,
               "max_depth": max_depth_list,
               "n_estimators": n_estimators_list}
```

```python
# GridSearchCV
model_xgb2 = GridSearchCV(estimator = xgboost.XGBClassifier(use_label_encoder = False,
                                                             random_state = 42),
                         param_grid = params_dict,
                         cv = 3,      
                         return_train_score = True)
model_xgb2.fit(X_train, y_train)
```

```python
# Getting the best parameters
print('params:', model_xgb2.best_params_)
```

```python
# Training xgboost on the best parameters.
model_xgb3 = xgboost.XGBClassifier(learning_rate = 0.02,
                                  max_depth = 3,
                                  n_estimators = 3000,
                                  random_state = 42)
model_xgb3.fit(X_train,
              y_train,
              eval_set = [(X_test, y_test)],
              verbose = True)
```
